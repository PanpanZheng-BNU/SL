1. 试证明
    1. $\hat{\beta}_0 + \hat{\beta}_1\bar{x} = \bar{y}$
    2. 当 $x_1 = \cdots = x_N$ 不成立时
    $$
    \hat{\beta}_{1} = \frac{\displaystyle \sum_{i=1}^{N}(x_{i}-\bar{x})(y_{i}-\bar{y})}{\displaystyle \sum_{i=1}^{N}(x_{i} - \bar{x})^{2}}
    $$
2. 若有一条曲线 $l$，其斜率为 $\hat{\beta}_{1}$，截距为 $\hat{\beta}_{0}$，求从 $\boldsymbol{x}-\boldsymbol{\bar{x}}$ 和 $\boldsymbol{y} - \boldsymbol{\bar{y}}$ 回归得出的新曲线 $l'$ 的斜率与截距
3. 试图可视化曲线 $l,l'$


```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import *


def min_sq(x, y):
    x_bar, y_bar = np.mean(x), np.mean(y)
    beta_1 = np.dot(x - x_bar, y - y_bar)/np.linalg.norm(x-x_bar) ** 2
    beta_0 = y_bar - beta_1 * x_bar
    return [beta_1, beta_0]
```

```python
N = 100
a = np.random.normal(loc=2, scale=1)
b = randn(1)
x = randn(N)
y = a * x + b + randn(N)

a1, b1 = min_sq(x, y)
xx = x - np.mean(x)
yy = y - np.mean(y)
a2, b2 = min_sq(xx, yy)
```

```python
x_seq = np.arange(-5, 5, 0.1)
y_pre = x_seq*a1 + b1
yy_pre = x_seq*a2 + b2
plt.scatter(x, y, c="black")
plt.axhline(y=0, c="black", lw=0.5)
plt.axvline(x=0, c="black", lw=0.5)
plt.plot(x_seq, y_pre, c="blue", label="before centering")
plt.plot(x_seq, yy_pre, c="orange", label="after centering")
```

    1. $\boldsymbol{Az} = \boldsymbol{0}\Leftrightarrow \boldsymbol{Bz} = \boldsymbol{0}$ 
    $$
    \begin{array}{ll}
      & \Rightarrow\\
      \because & \boldsymbol{Az} = \boldsymbol{0} \quad \boldsymbol{A} = \boldsymbol{B}^{\text{T}}\boldsymbol{B} \\
      \therefore & \boldsymbol{B}^{\text{T}}\boldsymbol{Bz} = \boldsymbol{0} \Rightarrow \boldsymbol{z}^{\text{T}}\boldsymbol{B}^{\text{T}}\boldsymbol{Bz} = \Vert \boldsymbol{Bz} \Vert^{2} = 0 \\
      \therefore & \boldsymbol{Bz} = \boldsymbol{0} \\\\
      & \Leftarrow\\
      \because & \boldsymbol{Bz} = \boldsymbol{0} \\
      \therefore & \boldsymbol{B}^{\text{T}}\boldsymbol{Bz} = \boldsymbol{0} \Rightarrow \boldsymbol{Az}= \boldsymbol{0} \\
    \end{array}
    $$
    2. 证明 $r(\boldsymbol{A}) = r(\boldsymbol{B})$ 
    $$
    \begin{array}{ll}
      \because & r(\boldsymbol{B}) = r(\boldsymbol{B^{\text{T}}}) \\
      \therefore & r(\boldsymbol{B}^{\text{T}}\boldsymbol{B}) \le r(\boldsymbol{B}) \\
    \end{array}
